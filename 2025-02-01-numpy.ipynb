```python
import cv2
```

```python
data = cv2.imread('pexels-mikebirdy-3729464.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
data.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(data)
```

```python
result = np.linalg.svd(data) # Singular value decomposition (SVD)
```

```python
result
```

```python
u, s, vt = np.linalg.svd(data) # Singular value decomposition
```

```python
u.shape
```

```python
s.shape
```

```python
vt.shape
```

```python
compressed = np.dot(u[:, :100], np.dot(np.diag(s[:100]), vt[:100, :])) # Reconstruct the image
```

```python
plt.imshow(compressed)
```
